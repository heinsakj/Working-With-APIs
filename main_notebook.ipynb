```python
crypto2 = requests.get('http://api.coincap.io/v2/markets')
crypto2
```

```python
c2 = crypto2.json()
c2
```

```python
pd.set_option('display.max_columns', None)
crypto_data2 = json_normalize(c2['data'])
crypto_data2
```

```python
# Find out what was traded in the last 24 hours the most
crypto_data2['baseSymbol'].value_counts().head(10)
crypto_data2.groupby(['baseSymbol','baseId', 'quoteId']).agg({'volumeUsd24Hr': sum}).sort_values(by='volumeUsd24Hr', ascending=False).reset_index().head(10)
```

- **baseSymbol**: most common symbol used to identify asset, base is asset purchased
- **baseID**: unique identifier for this asset, base is asset purchased
- **quoteID**: unique identifier for this asset, quote is asset used to purchase base
- **volumeUsd24Hr**: volume transacted on this market in last 24 hours

```python
crypto_data2.groupby(['type']).agg({'rateUsd':'count'}).reset_index()
```
